**1. Identifying Upgrades/Downgrades**

We've written a function `upgrader`. It checks if the user's first plan was a lower tier than the current plan they are rating. 

We've made a mistake using `.apply` to create a new `'grade'` column by looping through the rows of the DataFrame. Can you fix our mistake?

```python
def upgrader(row):
    if row['first_level'] == 'Free' and row['level'] != 'Free':
        output = 'Upgrader'
    elif row['first_level'] == 'Standard' and row['level'] == 'Premium':
        output = 'Upgrader'
    elif row['first_level'] == row['level']:
        output = 'Steady'
    else:
        output = 'Downgrader'
    return output

## YOUR SOLUTION HERE ##
csat['grade'] = csat.apply(upgrader,axis=1)

# show output
csat.head()
```

**2. Classifying Renewals**

We've written a function `renewed` that does a rough check to see if a customer's plan has renewed (to be more precise, we'd need to use date time functions to handle months and years of different lengths).

Apply our function to create a new `'renewed'` column.

```python
def renewed(row):
    if row['pay_freq'] == 'yearly' and row['days_subscribed'] > 365:
        output = True
    elif row['pay_freq'] == 'Monthly' and row['days_subscribed'] > 31:
        output = True
    else:
        output = False
    return output

## YOUR SOLUTION HERE ##
csat['renewed'] = csat.apply(renewed,axis=1)

# show output
csat[csat['renewed']].head()
```
